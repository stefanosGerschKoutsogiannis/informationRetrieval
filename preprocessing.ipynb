<h2> Join the two tables based on source_id <h2>

```python
def join_dataframes(df1: pd.DataFrame, df2: pd.DataFrame, on: str, how: str):
    table = df1.merge(df2, on=on, how=how)
    return table
```

```python
# work on the copies, not the original data
table = join_dataframes(papers_copy, authors_copy, 'source_id', 'inner')
table
```

```python
table.info()
```

```python
table.isnull().sum()
```

```python
drop_nan = table.dropna()
drop_nan.info()
```

```python
drop_nan.isnull().sum()
```

```python
drop_dupl = drop_nan.drop_duplicates()
```

```python
pd.options.mode.copy_on_write = True 
drop_dupl['authors'] = drop_dupl['first_name'] + ' ' + drop_dupl['last_name']
```

```python
drop_dupl.info()
```

```python
drop_names = drop_dupl.drop(columns=['first_name', 'last_name', 'institution', 'source_id'])
drop_names
```

```python
remove_commas = drop_names.replace(',', ' ', regex=True)
```

```python
final = remove_commas.groupby('title').agg({
    'year': 'first',
    'title': 'first',
    'abstract': 'first',
    'full_text': 'first',
    'authors': lambda x: '\n'.join(x)
    })
```

```python
final
```

```python
# this is stored, => title, year,
final.reset_index(drop=True, inplace=True)
final
```

<h2> Save the new dataset </h2>

<h2> Make csv files and them make them txt</h2> 

```python
outpup_directory = r'C:\Users\STEVE\Desktop\searchEngine\paperWorld\src\main\resources\corpus'

data = pd.read_csv(r'C:\Users\STEVE\Desktop\searchEngine\paperWorld\src\main\resources\original\papers_authors.csv')


for index, row in data.iterrows():
    filename = f'entry_{index+1}.csv'
    
    row_data = ','.join(row.astype(str))
    
    with open(outpup_directory+filename, 'w', encoding='utf-8') as file:
        file.write(row_data)
    
```
